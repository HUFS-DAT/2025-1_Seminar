```python
from tbparse import SummaryReader
import matplotlib.pyplot as plt
import math
```

```python
# 로그 파일 경로 (runs/exp1 등)
log_dir = '../tensorboard/experiment_3'
reader = SummaryReader(log_dir)

df = reader.scalars  # 전체 로그가 들어 있는 DataFrame
print(df.head())


# 상위 그룹 리스트
groups = ['Train/', 'Val/', 'LearningRate/']

for group in groups:
    group_tags = [tag for tag in df['tag'].unique() if tag.startswith(group)]
    
    if not group_tags:
        continue
    
    # 서브플롯 행/열 구성
    n_tags = len(group_tags)
    n_cols = 2  # 가로 2개
    n_rows = math.ceil(n_tags / n_cols)
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(10, 4 * n_rows))
    axes = axes.flatten()  # 1D 배열로 만들어 인덱싱 쉽게
    
    for idx, tag in enumerate(group_tags):
        sub_df = df[df['tag'] == tag]
        axes[idx].plot(sub_df['step'], sub_df['value'])
        axes[idx].set_title(tag)
        axes[idx].set_xlabel('Epoch')
        axes[idx].set_ylabel('Value')
        axes[idx].grid(True)
    
    # 남는 subplot 지우기
    for j in range(len(group_tags), len(axes)):
        fig.delaxes(axes[j])
    
    fig.suptitle(f"{group.strip('/')}", fontsize=16)
    plt.tight_layout()
    plt.subplots_adjust(top=0.9)
    plt.show()
```
